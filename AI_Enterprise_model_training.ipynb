```python
data.columns=['Area', 'Perimeter', 'Compactness', 'KernelLength','KernelWidth', 'AsymmetryCoeff','KernelGroove' , 'Category']
data.head()
```

```python
data["Category"].replace({1 :"Kama" , 2 :"Rosa" , 3 :"Canadian"} , inplace = True)
data.head(-5)
```

```python
import matplotlib.pyplot as plt
```

```python
data.Category.value_counts(normalize=True).plot.barh()    
plt.show()    #Plotting the bar graph of percentages seed categories
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(data.corr(), annot=True, square=True ,ax=ax, cmap = 'Reds')
plt.show()    #Correlation Matrix
```

```python
sns.relplot(x='Perimeter', y='Compactness', hue='Category', data=data);
#ScatterPlot
```

```python
sns.boxplot(data.Category, data.Compactness)
#Plotting boxplot of Category for Compactness
```

```python
x= data.drop('Category', axis=1)
x
```

```python
y= data['Category']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, random_state=2, test_size = 0.25)
```

```python
model = DecisionTreeClassifier(min_samples_split=50, criterion='gini')
```

```python
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
print("The accuracy of the model is: " + str(accuracy_score(y_test, y_predict)*100) + ' %')
```

```python
import pickle
```

```python
filename='model.pkl'
outfile= open(filename, 'wb')
pickle.dump(model, outfile)
outfile.close()
```
